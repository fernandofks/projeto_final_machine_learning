## Separando os dados

```python
# Separando a variável alvo das demais

y = data_of_interest['log_salary'].copy()
X = data_of_interest.drop(columns=['salary_in_usd', 'log_salary']).copy()
X = X.astype('float')
```

```python
# Avaliando a formato delas

X.shape, y.shape
```

```python
# Separando conjunto em treinamento e test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Separando conjunto em treinamento e validação

X_train_valid, X_test_valid, y_train_valid, y_test_valid = train_test_split(X_train, y_train, test_size=0.33)
```

## Modelo Ridge

```python
# Importando bibliotecas necessárias

import math
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error

# Criando Modelo com alpha

model = Ridge(alpha=1e-3)

# Treinando modelo

model.fit(X_train_valid, y_train_valid)

# Fazendo a predição

y_pred = model.predict(X_test_valid)

# Obtendo RMSE do conjunto de validação

RMSE = math.sqrt(mean_squared_error(y_test_valid, y_pred))
RMSE
```

```python
# Verificando a influencia de cada feature a partir de seus pesos

list(zip(X_test_valid.columns, (((10**model.coef_)-1)*100)))
```

```python
# Comparação entre valores reais e preditos

plt.scatter(y_test_valid, y_pred)
```

```python
# Calculando erros

erro = y_pred - y_test_valid

plt.scatter(y_test_valid, erro)
plt.xlabel('salario')
plt.ylabel('Erro [$]')
```

```python
# Cálculo do erro percentual

erro_percentual = (10**erro) - 1
```

```python
plt.scatter(y_test_valid, erro_percentual)
plt.xlabel('salario ')
plt.ylabel('Erro [%]')
```

## Modelo Ridge com polinomial de grau 2

```python
# Importando bibliotecas necessárias

from sklearn.preprocessing import PolynomialFeatures

# Criando Polynomial Feature com grau 2

transformer = PolynomialFeatures(degree=2, include_bias=False)

# Aplicando aos dados

X_train_valid_transf = transformer.fit_transform(X_train_valid)
```

```python
# Observando a formato dos dados após Polynomial Features

X_train_valid_transf.shape
```

```python
# Criando modelo Ridge para dados após Polynomial Features (PF)

alpha_PF = 1e-3
model_PF = Ridge(alpha=alpha_PF)

model_PF.fit(X_train_valid_transf, y_train_valid)
```

```python
# Aplicando Polynomial Features para conjunto X de validação

X_test_valid_transf = transformer.transform(X_test_valid)

# Predizendo dados de com conjunto de validação

y_pred_FT = model_PF.predict(X_test_valid_transf)
```

```python
# Verificando RMSE novo

RMSE_FT = math.sqrt(mean_squared_error(y_test_valid, y_pred_FT))
RMSE_FT
```

```python
# Comparação entre valores reais e preditos

plt.scatter(y_test_valid, y_pred_FT, alpha=0.3)
```

```python
# Calculando erros

erro_FT = y_pred_FT - y_test_valid

plt.scatter(y_test_valid, erro_FT)
plt.xlabel('salario')
plt.ylabel('Erro [$]')
```

```python
# Cálculo do erro percentual

erro_percentual_FT = (10**erro_FT) - 1
```

```python
plt.scatter(y_test_valid, erro_percentual_FT)
plt.xlabel('salario')
plt.ylabel('Erro [%]')
```
